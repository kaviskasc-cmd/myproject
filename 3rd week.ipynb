## types of argument
required or positional - should give the required input to perform
default - where there is default input it will accept or if there is a given input it will ignore default and take only given input
keyword - when there is mismatch in the input it will take as it is or we can assign a key to the input we are giving
variable length - when there is only one argument in the variable but has 2 or 3 inputs it will throw an error. we can give * (arbitory) to assign multiple input in a single argument

```python
#required or positional argument
def details(name,des):
    print(f"hai {name} welcome to food company")
    print(f"congrats .. you are selected as a {des}")
details("Sneha","food inspector")
```

```python
#default argument
def details(name = "riya",des="data analyst"):
    print(f"hai {name} welcome to xyz company")
    print(f"congrats .. you are selected as a {des}")
details()
```

```python
#variable length
def course(list):
    print(list)
course('cds','cda','ai')
```

```python
#variable length- when we give *(arbitory) in the argument it will take n number of input to perform
def course(*list):
    print(list)
course('cds','cda','ai')
```

```python
# variable length- when we need it in the list use for 
def course(*list):
    for i in list:
        print(i)
course('cds','cda','ai')
```

```python
#keyword - we can assign the key to the input we are giving so it assigns accordingly
def company(des,name):
    print(f" Hi! {name} please collect your id card")
    print(f" congrats! you have destination change to {des}")
company(name="Geetha",des="Manager")
```

## Anonymous function

```python
#function without the name is called anonymous we have to define with lambda keyword
#syntax
lambda argumentlist:expression
```

```python
res=lambda n1,n2,n3:n1+n2+n3
print(res(23,45,56))
```

```python
maximum=lambda num1,num2:num1 if num1>num2 else num2
print(maximum(345,567))
```

```python
minimum=lambda no1,no2:no1 if no1<no2 else no2
print(minimum(34,56))
```

```python
combine=lambda a,b:a+b
print(combine("sam","sung"))
```

```python
result=lambda a="data":a.upper()
print(result())
```

```python
result=lambda a="analyst":a.lower()
print(result())
```

```python
result=lambda a:"even" if a%2==0 else "odd"
print(result(4))
```

```python
result=lambda a:a**2
print(result(8))
```

# built in functions
filter(function,sequence)
map(function,sequence)
reduce(function,sequence)

```python
odd_no=filter(lambda num:num%2!=0,(34,89,76,9))
print(set(odd_no))
```

```python
def even_no(no):
    if no%2==0:
        return True
    else:
        return False
ans=filter(even_no,[34,32,2,7,9])
print(list(ans))
```

```python
name=filter(lambda a:a.startswith ('a'),{'ankita','anan','arivu','liya','riya','arun'})
print(set(name))
```

```python
name=map(lambda a:a.capitalize(),('ankita','anan','arivu','liya','riya','arun'))
print(set(name))
```

```python
name=filter(lambda x:x.endswith ('tha'),{'neetha','geetha','meena'})
print(set(name))
```

```python
#map (function,sequence)
t1=(1,2,3,4)
t2=(3,5,6,7)
num=map(lambda n1,n2:n1*n2,t1,t2)
print(list(num))
```

```python
f_name=('Sweta','binny','tom')
l_name=('jain','joseph','thomas')
name=map(lambda a,b:a + " "+ b,f_name,l_name)
for n in name:
    print(f"hai {n} welcome to the class")
```

```python
#reduce(function,sequence)
from functools import reduce
red=reduce(lambda n1,n2:n1+n2,[90,80,70,60])
print(red)
```

```python
fruits=['apple','orange','watermelon','custardapple']
longest_word=reduce(lambda x,y:x if len(x)>len(y) else y, fruits)
print(longest_word)
```
